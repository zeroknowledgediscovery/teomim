### **Made with 30 10% samples of 2888 POS + 2888 NEG set**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


import warnings
warnings.filterwarnings("ignore")
```

```python
from matplotlib.ticker import MaxNLocator
GLOBAL_FIGSIZE = 4
LABEL_SIZE = 13

GLOBAL_FONTSIZE = 22
plt.style.use('fivethirtyeight')
```

```python
BAREBONES_AUC = pd.read_csv(
    f"BAREBONES_AUC.csv"
)
```

```python
plt.figure(figsize = (GLOBAL_FIGSIZE, GLOBAL_FIGSIZE))

sns.distplot(BAREBONES_AUC['auc_zcor'], 
             color = 'k', label = 'ZCoR AUC', 
             bins = 6)
sns.distplot(BAREBONES_AUC['auc_digitaltwin'], 
             color = 'r', label = 'Phantom AUC',
            bins = 6)

plt.legend(loc = "upper right", 
           fontsize = GLOBAL_FONTSIZE, 
           bbox_to_anchor=(1.04, 1))

plt.xlabel("AUC", fontsize = GLOBAL_FONTSIZE)
plt.ylabel("Density", fontsize = GLOBAL_FONTSIZE)

plt.xticks(fontsize=GLOBAL_FONTSIZE)
plt.yticks(fontsize=GLOBAL_FONTSIZE)

plt.grid()
plt.legend()
plt.savefig(
    "TASK_PLOTS/AUC_IMPROVEMENT_HISTOGRAM.png",
    dpi = 300,
    bbox_inches = 'tight'
)
plt.show
```
