### **Made with 30 10% samples of 2888 POS + 2888 NEG set**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


import warnings
warnings.filterwarnings("ignore")
```

```python
BAREBONES_AUC = pd.read_csv(
    f"BAREBONES_AUC.csv"
)
```

```python
plt.figure(figsize = (7,4))

sns.distplot(BAREBONES_AUC['auc_zcor'], 
             color = 'k', label = 'ZCoR AUC', 
             bins = 6)
sns.distplot(BAREBONES_AUC['auc_digitaltwin'], 
             color = 'r', label = 'ZCoR + Digital Twins AUC',
            bins = 6)
plt.grid()
plt.legend()
plt.savefig(
    "TASK_PLOTS/AUC_IMPROVEMENT_HISTOGRAM.png",
    dpi = 300,
    bbox_inches = 'tight'
)
plt.show
```
